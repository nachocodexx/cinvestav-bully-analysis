# Throughput[LEADER]

```python
COLUMNS_RENAME = {
    "D0":"OPERATION_ID",
    "D1":"FILE_ID",
    "D2":"METHOD",
    "D3":"COMPRESSION_TIME",
    "D4":"SIZE_IN",
    "D5":"SIZE_OUT",
    "D6":"MB_PER_SECOND",
    "D7":"EXPERIMENT_ID"
}
COL_UP = {"D0":"OPERATION_ID",
          "D1":"FILE_ID",
          "D2":"USER_ID", 
          "D3":"URL",
          "D4":"RF",
          "D5":"EXPERIMENT_ID"
         }

# path = './wl_peers_3/sn-2-20210623T211729.txt'
# path = './wl_peers_6/sn-5-20210623T225109.txt'
path = './data/sn-11-20210624T030615.txt'
def getData(path):
    sn_data = pd.read_csv(path,header=None,sep=" ",names=names) 
    cmd_col = sn_data["COMMAND"]
    sn_data_filtered = sn_data[(cmd_col=="UPLOAD_FILE")  | 
            (cmd_col=="SAVE_FILE_INIT") | 
            (cmd_col=="SAVE_FILE_DONE") | 
            (cmd_col=="COMPRESSION_INIT") |
            (cmd_col=="COMPRESSION_DONE")
           ]

    cmd_col_fil=sn_data_filtered["COMMAND"]
    # COMPRESSION OPERATIONS 
    compression_ops  = sn_data_filtered[cmd_col_fil=="COMPRESSION_DONE"]
    # EXPERIMENT DATA
    compression_done_data  = compression_ops.rename(columns = COLUMNS_RENAME).reset_index(drop=True)
    upload_files           = sn_data_filtered[cmd_col_fil=="UPLOAD_FILE"].rename(columns=COL_UP).drop(["D6","D7"],axis=1).reset_index(drop=True)
    return sn_data_filtered,compression_done_data,upload_files
# sn_data_filtered,compression_done_data,upload_data = getData(path)
```

```python
def getSTandThroughput(path):
    sn_data_filtered,compression_done_data,upload_data = getData(path)
    STs = []
    SMs = []
    for i in range(MAX_EXPERIMENTS):
        ei = compression_done_data[compression_done_data["EXPERIMENT_ID"]==i]
#         end = ei["TIME"]
        end = ei["RELATIVE_TIME"]
        eupi = upload_data[upload_data["EXPERIMENT_ID"]==i]
#         start = eupi["TIME"]
        start = eupi["RELATIVE_TIME"]
        serviceTimes = (end-start)
        ST_avg = serviceTimes.mean()
        SM     = serviceTimes.sum()
#         print(serviceTimes.unique())
        STs.append(ST_avg)
        SMs.append(SM)
    #     print(V)
    STs = np.array(STs)
    SMs = np.array(SMs)
    return STs.mean(),SMs.mean()
```

```python
path = './wl_peers_3/sn-2-20210623T211729.txt'
ST,SM = getSTandThroughput(path)
print("ST: {} SM = {} Throughput: {}".format(ST,SM,VOLUME_GB/(SM/1000)))
```

```python
rep_cols = {"D0":"OPERATION_ID","D1":"FILE_ID","D2":"EXPERIMENT_ID"}
repd_cols = {"D0":"OPERATION_ID","D1":"FILE_ID","D2":"VOLUME","D3":"EXPERIMENT_ID"}
def getNodeData(**kwargs):
    folder_name = kwargs.get("folder_name","./data")
    leader      = kwargs.get("leader","sn-5")
    files = os.listdir(folder_name)
    def filterFiles(x):
        if(x.startswith(leader)):
            return False
        return x.startswith("sn")

    files = list(filter(filterFiles,files))

    replica_service_time = []
    throughputs          = []
    nodes_data           = {}
    for file in files:
        data = pd.read_csv(folder_name+"/"+file,header=None,sep=" ",names=names)
    #     node_index = file.split("-")[1]
        node_id    =data["NODE_ID"][0]
        if not (node_id in nodes_data):
            nodes_data[node_id] = {
                'SIMULATION_TIMES':[],
                'AVG_SERVICE_TIMES':[], 
                'VOLUMES':[],
                'THROUGHPUTS':[],
                'AVG_SERVICE_TIME':0,
                'AVG_THROUGHPUT':0
            }
            
        rep_data  = data[data["COMMAND"]=="REPLICATION"].rename(columns=rep_cols).drop(["D3","D4","D5","D6","D7"],axis=1).reset_index(drop=True)
        rep_data["EXPERIMENT_ID"]  = pd.to_numeric(rep_data["EXPERIMENT_ID"])
        repd_data = data[data["COMMAND"]=="SAVE_REPLICA_DONE"].rename(columns=repd_cols).drop(["D4","D5","D6","D7"],axis=1).reset_index(drop=True)
        repd_data["EXPERIMENT_ID"]  = pd.to_numeric(repd_data["EXPERIMENT_ID"])
#         print(repd_data["EXPERIMENT_ID"].unique())
#         print(rep_data[rep_data["EXPERIMENT_ID"]==0])
#         break
        for i in range(MAX_EXPERIMENTS):
            exp_data        = [
                             rep_data[rep_data["EXPERIMENT_ID"]==i],
                             repd_data[repd_data["EXPERIMENT_ID"]==i]
                            ]
            end             = exp_data[1]["TIME"]
            start           = exp_data[0]["TIME"]
#             print(exp_data[0])
#             print(end)
    #         
            service_times   = end-start
            simulation_time = service_times.sum()
            st_avg          = service_times.mean()
    #         
            volume          = pd.to_numeric(exp_data[1]["VOLUME"]).sum()
#             print(service_times)
            throughput      = (volume/1000000)/(simulation_time/1000)
            nodes_data[node_id]["VOLUMES"].append(volume)
            nodes_data[node_id]['SIMULATION_TIMES'].append(simulation_time)
            nodes_data[node_id]['AVG_SERVICE_TIMES'].append(st_avg)
            nodes_data[node_id]['THROUGHPUTS'].append(throughput)
        nodes_data[node_id]["SIMULATION_TIMES"] = np.array(nodes_data[node_id]["SIMULATION_TIMES"])
        nodes_data[node_id]["AVG_SIMULATION_TIME"] = nodes_data[node_id]["SIMULATION_TIMES"].mean()
        nodes_data[node_id]["VOLUMES"] = np.array(nodes_data[node_id]["VOLUMES"])
        nodes_data[node_id]["AVG_VOLUME"] = nodes_data[node_id]["VOLUMES"].mean()
        nodes_data[node_id]["AVG_SERVICE_TIMES"] = np.array(nodes_data[node_id]["AVG_SERVICE_TIMES"])
        nodes_data[node_id]["AVG_SERVICE_TIME"] = nodes_data[node_id]["AVG_SERVICE_TIMES"].mean()
        nodes_data[node_id]["THROUGHPUTS"] = np.array(nodes_data[node_id]["THROUGHPUTS"])
        nodes_data[node_id]["AVG_THROUGHPUT"] = nodes_data[node_id]["THROUGHPUTS"].mean()
#         nodes_data[node_id]["THROUGHPUTV2"] = (nodes_data[node_id]["VOLUMES"].mean()/1000000)/(nodes_data[node_id]["SIMULATION_TIMES"].mean()/1000)
    service_times = []
    throughputs   = []
    simulation_times =[]
    for node_id,data in nodes_data.items():
        st= data["AVG_SERVICE_TIME"]
        tp = data["AVG_THROUGHPUT"]
        sm = data["AVG_SIMULATION_TIME"]
        print(node_id+" THROUGHPUT: "+str(tp))
        print(node_id+" SERVICE_TIME: "+str(st))
        print(node_id+" SIMULATION_TIME: "+str(sm))
        service_times.append(st)
        throughputs.append(tp)
        simulation_times.append(sm)
        print("_"*50)
    service_times = np.array(service_times)
    throughputs = np.array(throughputs)
    simulation_times= np.array(simulation_times)
    nodes_data["SERVICE_TIME"]=service_times.mean()
    nodes_data["THROUGHPUT"]=throughputs.mean()
    nodes_data["SIMULATION_TIME"]=simulation_times.mean()
#     nodes_data["SIMULATION_TIME"]=simulation_times.sum()
#     for node_id, values in nodes_data.items():
# #         V = (values["AVG_VOLUME"]/100000)
# #         real_t = (data["AVG_VOLUME"]/1000000)/(nodes_data["SIMULATION_TIME"]/1000)
#         print(values)
#         print("_"*50)
#     (service_times.mean(),throughputs.mean())
    return nodes_data 
# print(nodes_data)
```
